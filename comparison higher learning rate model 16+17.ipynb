```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
test_dataset_16 = pd.read_parquet("model_16_comparison_test.parquet")
test_dataset_17 = pd.read_parquet("model_17_comparison_test.parquet")
```

```python
percential_deviation = lambda x, y: abs(((y-x)/x)*100)
```

```python
test_dataset_16["percential deviation"] = [percential_deviation(*a) for a in tuple(zip(test_dataset_16["concentration"], test_dataset_16["prediction"]))]
test_dataset_17["percential deviation"] = [percential_deviation(*a) for a in tuple(zip(test_dataset_17["concentration"], test_dataset_17["prediction"]))]
```

```python
print(test_dataset_16.describe())
print(test_dataset_17.describe())
```

```python
sns.scatterplot(test_dataset_16["concentration"], test_dataset_16["percential deviation"])
plt.show()
sns.scatterplot(test_dataset_17["concentration"], test_dataset_17["percential deviation"])
```

```python
test_dataset_16.boxplot(column="percential deviation")
plt.show()
test_dataset_17.boxplot(column="percential deviation")
plt.show()
```
